```
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```

x_train_pos=np.load("/content/drive/My Drive/Numpy_Files/5000_pos/5000_pos_ENAC_nor.npy")
x_train_neg=np.load("/content/drive/My Drive/Numpy_Files/5000_neg/5000_neg_ENAC_nor.npy")
x_test_pos=np.load("/content/drive/My Drive/Numpy_Files/1000_pos/1000_pos_ENAC_nor.npy")
x_test_neg=np.load("/content/drive/My Drive/Numpy_Files/1000_neg/1000_neg_ENAC_nor.npy")
y_train_pos=np.tile(1,5289)
y_train_neg=np.tile(0,5289)
y_test_pos=np.tile(1,1000)
y_test_neg=np.tile(0,1000)
print(x_train_pos.shape)
print(y_train_pos.shape)
print(x_test_pos.shape)
print(y_test_pos.shape)
```

```
seed=42
np.random.seed(seed)
```

```
x_training_pos, x_val_pos, y_training_pos, y_val_pos = train_test_split(x_train_pos, y_train_pos, test_size=0.20, random_state=seed)

x_training_neg, x_val_neg, y_training_neg, y_val_neg = train_test_split(x_train_neg, y_train_neg, test_size=0.20, random_state=seed)
print(x_training_pos.shape)
print(x_val_pos.shape)
```

```
X_train = np.concatenate((x_training_pos, x_training_neg), axis=0)
Y_train = np.concatenate((y_training_pos, y_training_neg))

X_val = np.concatenate((x_val_pos, x_val_neg), axis=0)
Y_val = np.concatenate((y_val_pos, y_val_neg))
print(X_train.shape)
```

```
X_test = np.concatenate((x_test_pos, x_test_neg))
Y_test = np.concatenate((y_test_pos, y_test_neg))
print(X_test.shape)
print(Y_test.shape)
```

```
c = .25
clf = SVC(C=c,  random_state = seed)
```

```
clf.fit(X_train, Y_train)
```

```
Y_val_pred = clf.predict(X_val)
```

```
print(Y_val.shape)
print(Y_val_pred.shape)
```

```
val_accuracy=accuracy_score(Y_val,Y_val_pred)
print(val_accuracy)
```

```
Y_test_pred=clf.predict(X_test)
```

```
test_accuracy=accuracy_score(Y_test,Y_test_pred)
print(test_accuracy)
```

```
Y_train_pred=clf.predict(X_train)
```

```
train_accuracy=accuracy_score(Y_train,Y_train_pred)
print(train_accuracy)
```
